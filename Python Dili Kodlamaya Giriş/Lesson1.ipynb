## Conda, Miniconda ve Anaconda Arasındaki Farklar

Python geliştirme ortamı kurarken karşımıza çıkan üç temel seçenek:

### Conda
- Paket ve ortam (environment) yöneticisi
- Sadece Python değil, diğer dilleri de destekler
- Bağımlılıkları otomatik olarak yönetir

### Miniconda
- Sadece conda + birkaç temel paket içerir
- Minimal kurulum (yaklaşık 400 MB)
- İhtiyaç duyduğunuz paketleri sonradan yüklersiniz
- Disk alanı tasarrufu sağlar
- Daha hızlı kurulum

### Anaconda
- İçinde conda'nın da olduğu "büyük" dağıtım paketi
- 150+ popüler veri bilimi paketi önceden yüklü gelir
- Jupyter, Spyder, VS Code gibi IDE'ler dahil
- Yaklaşık 3 GB disk alanı gerektirir
- Başlangıç için kullanışlı (çok sayıda paket yüklü)

**Hangi birini seçmeliyim?**
- Yeni başlıyorsanız ve disk alanınız varsa: **Anaconda**
- Minimal kurulum istiyorsanız: **Miniconda**
- Sadece paket yönetimi için: **Conda** (Miniconda ile gelir)
